# Create new dataframe for project information

### Non-mandatory projects

```python
#option based
la_NONmandatory_proj = la[(la.scholarly_project == 1)  & (la.Compulsory_YES_ID == "2 - No")]

print(len(la_NONmandatory_proj.College_ID.unique()))

la_NONmandatory_proj
```

```python
#write in proj options for those with non-mandatory projects

for ids in la_NONmandatory_proj ["College_ID"]:
    
    #get index
    index = la_NONmandatory_proj[la_NONmandatory_proj['College_ID']== ids].index[0]
    
        
    if la_NONmandatory_proj.loc[index, "project_category_ID"] == '1 - Only option': 
        colleges.loc[colleges['College'] == ids, 'Project Requirements'] = "Optional"
        
    elif la_NONmandatory_proj.loc[index, "project_category_ID"] == '2 - Options based system': 
        colleges.loc[colleges['College'] == ids, 'Project Requirements'] = "Optional (a)"
        
    else:
        colleges.loc[colleges['College'] == ids, 'Project Requirements'] = "Optional (b)"

    
print(colleges['Project Requirements'].value_counts(dropna=False))
colleges.head()
```

## Experienced Supvervisor Required

```python
#get a dataframe of all the scholarly projects
la_proj = la[(la.scholarly_project == 1)]


#reindex
la_proj = la_proj.reset_index(drop=True)


#get the values in the LEARNINGA_experienced_supervisor_ID column
print(la_proj.LEARNINGA_experienced_supervisor_ID.unique())

print(la_proj["LEARNINGA_experienced_supervisor_ID"].value_counts(dropna=False))


```

```python
la_proj[la_proj["LEARNINGA_experienced_supervisor_ID"]== '1 - Supervisor NOT required and/or mentioned']
```

```python
#write in proj options for those with non-mandatory projects

#which df
df = la_proj

#which column to enter data
col_in = "Experienced supervisor specified"

#which column to get data out of
col_out = "LEARNINGA_experienced_supervisor_ID"


for ids in df ["College_ID"]:
    
    #get index
    index = df[df['College_ID']== ids].index[0]
       
    if df.loc[index, col_out] == '3 - Supervisor required but no mention of RESEARCH experience': 
        colleges.loc[colleges['College'] == ids, col_in] = "No"
        
    if df.loc[index, col_out] == '2 - Supervisor required but no details given': 
        colleges.loc[colleges['College'] == ids, col_in] = "No"

        
for ids in df ["College_ID"]:
    
    #get index
    index = df[df['College_ID']== ids].index[0]

    if df.loc[index, col_out] == '1 - Supervisor NOT required and/or mentioned': 
        colleges.loc[colleges['College'] == ids, col_in] = "NR"


        
for ids in df ["College_ID"]:
    
    #get index
    index = df[df['College_ID']== ids].index[0]
       
    
    if df.loc[index, col_out] == '4 - Experienced research supervisor ENCOURAGED': 
        colleges.loc[colleges['College'] == ids, col_in] = "Yes*"
        
        
    
for ids in df ["College_ID"]:
    
    #get index
    index = df[df['College_ID']== ids].index[0]
    
    if df.loc[index, col_out] == '5 - Experienced research supervisor MANDATORY': 
        colleges.loc[colleges['College'] == ids, col_in] = "Yes"
    
    
    
    
print(colleges[col_in].value_counts(dropna=False))
colleges.head()
```
